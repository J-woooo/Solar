```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv("569. LH서울공릉 행복주택 발전량.csv", encoding='euc-kr')
# 인버터 개수, 날짜, 시간, 인버터 출력, 누적 발전량
df = df[['pow_id', 'pow_date', 'pow_time', 'pow_cap', 'pow_totpower']]
df.head()
```

```python
# 발전량 column추가 기본값:0
df['power'] = 0
```

```python
# 시간 변경&발전량 계산. 겁나오래걸림
for i in range (df.shape[0]):
    df['pow_time'][i] = df['pow_time'][i][:-6]
    if i != (df.shape[0]-1):
        df['power'][i] = df['pow_totpower'][i+1] - df['pow_totpower'][i]
    if(i%1000 == 0):
        print(i)
```

```python
df.to_csv("569. LH서울공릉 행복주택 발전량(수정).csv")
```

```python
df2 = pd.read_csv("615. 보듬어린이집 발전량(수정).csv", encoding='euc-kr')
df2 = df2[['pow_id', 'pow_date', 'pow_time', 'pow_cap', 'power']]
df2.head()
```

```python
power_cap = df2['pow_cap'][0]
power_cap
```

```python
# pow_date+pow_time 별로
date_unique = df2['pow_date'].unique().tolist()
time_unique = df2['pow_time'].unique().tolist()
```

```python
date_list = []
for date in date_unique:
    for time in time_unique:
        date_list.append(date)
```

```python
data = {'pow_date': date_list,
       'pow_time':time_unique*len(date_unique)}
final_df = pd.DataFrame(data)
final_df['power_cap'] = power_cap
final_df['power'] = 0
```

```python
for index, row in final_df.iterrows():
    d = row['pow_date']
    t = row['pow_time']
    power_list = df2[(df2['pow_date']==d) & (df2['pow_time']==t)]['power']
    total_power = 0
    for power in power_list:
        total_power += power
    final_df['power'][index] = total_power
```

```python
final_df.to_csv("569. LH서울공릉 행복주택 발전량(최종수정).csv")
```
