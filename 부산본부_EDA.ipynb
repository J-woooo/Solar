### 최소자승법 OLS를 적용한 회귀분석

```python
from statsmodels.formula.api import ols
```

```python
# 필요없는 column을 제거하고 회귀분석 
X = df.drop(['일시', '총량(kw)'], axis='columns')
y = df['총량(kw)']
res = ols('y ~ X', data=df).fit()
```
