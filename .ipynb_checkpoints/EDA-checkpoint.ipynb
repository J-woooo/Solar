```python
import os

plt.rc('font', family='Malgun Gothic')
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

#ax.plot(df_test1['년월일'], df_test1['총량(kw)'], label = '총량')
ax.plot(df_test1['년월일'], df_test1['최저기온(°C)'], label = '최저기온(°C)')
ax.plot(df_test1['년월일'], df_test1['최고기온(°C)'], label = '최고기온(°C)')
ax.plot(df_test1['년월일'], df_test1['일강수량(mm)'], label = '일강수량(mm)')
ax.plot(df_test1['년월일'], df_test1['평균 풍속(m/s)'], label = '평균 풍속(m/s)')
ax.plot(df_test1['년월일'], df_test1['평균 상대습도(%)'], label = '평균 상대습도(%)')
ax.plot(df_test1['년월일'], df_test1['가조시간(hr)'], label = '가조시간(hr)')
ax.plot(df_test1['년월일'], df_test1['합계 일사량(MJ/m2)'], label = '합계 일사량(MJ/m2)')
ax.plot(df_test1['년월일'], df_test1['평균 전운량(1/10)'], label = '평균 전운량(1/10)')

ax.set_title("13년도 추이")
ax.set_xlabel("년월")
ax.set_ylabel("변화량")

ax.legend()
plt.show()
```

```python
df_test2 = df[['년월일', '총량(kw)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '평균 상대습도(%)', '가조시간(hr)', '합계 일사량(MJ/m2)', '평균 전운량(1/10)']]
df_test2 = df_test2.fillna(0)
df_test2.head()
#df_test2 = df_test2.loc[0:100, :]

x = [df_test2['최저기온(°C)'], df_test2['최고기온(°C)'], df_test2['일강수량(mm)'], df_test2['평균 풍속(m/s)'], df_test2['평균 상대습도(%)'], df_test2['가조시간(hr)'], df_test2['합계 일사량(MJ/m2)'], df_test2['평균 전운량(1/10)']]
labels = ['최저기온(°C)', '최고기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '평균 상대습도(%)', '가조시간(hr)', '합계 일사량(MJ/m2)', '평균 전운량(1/10)']
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

ax.boxplot(x,labels = labels)
plt.show()
```
